```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("first_order_analysis.csv")

print(df["days_to_first_order"].describe())

df.to_csv("days_to_first_order_clean.csv", index=False)

plt.hist(df["days_to_first_order"], bins=20, edgecolor="black")
plt.title("Distribution of Days to First Order")
plt.xlabel("Days")
plt.ylabel("Number of Customers")
plt.show()
```
